```python
type(text9)
```

```python
tokens = [i for i in text9.tokens if i not in """.,:;'"!?[]"""]
```

```python
counter = Counter(tokens)
```

```python
counter.most_common(20)
```

```python
tokens = []
matches = 0
for token in text9.tokens:
    for char in """.,:'";!?[]-""":
        if char in token:
            matches += 1
    if matches < 1:
        tokens.append(token)
    matches = 0
```

```python
counter = Counter(tokens)
counter.most_common(20)
```
